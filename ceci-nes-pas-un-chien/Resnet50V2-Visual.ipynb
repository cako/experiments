```python
import tensorflow as tf
import visualkeras as vk

from PIL import ImageFont
font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf", size=160)
```

```python
model = tf.keras.applications.ResNet50V2(include_top=True)
```

```python
vk.layered_view(model, legend=True, font=font, scale_xy=20, to_file='resnet50v2.png')
```
